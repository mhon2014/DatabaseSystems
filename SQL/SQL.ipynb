### Functions used to create and execute queries

````python
def batch_execute_ddl(conn, file):
    cursor = conn.cursor()
    ddl_file = open(file)
    sql = ddl_file.read()
    
    # print(sql)
    # multi = True for multple statements
    for result in cursor.execute(sql, multi=True):
        if result.with_rows:
            print(f"Rows produced: {result.statement}")
            print(result.fetchall())
        else: 
            print(f"number of rows affected by statement {result.statement}: {result.rowcount}")
    
    conn.close()

def displaytables(conn):
    cursor = conn.cursor()
    tables = ['branch', 'account', 'depositor', 'customer', 'loan', 'borrower']
    
    for table in tables:
        cursor.execute('SHOW CREATE TABLE ' + table ) 
        result = cursor.fetchone()
        
        string = ''.join(result)
        
        # string = \'''\' + string + \'''\'
        # print(string + "\n")
        test = """('CREATE TABLE `branch` (\n  `branch_name` varchar(255) NOT NULL,\n  `branch_city` varchar(255) DEFAULT NULL,\n  `assets` float DEFAULT NULL,\n  PRIMARY KEY (`branch_name`)\n) ENGINE=InnoDB DEFAULT CHARSET=utf8mb4 COLLATE=utf8mb4_0900_ai_ci')"""
        
        # print(result)
        
#         #Honestly no idea how this works...is it suppose to take a variable in or just strings???
        display(Markdown(f"""
        ```sql
        {test}
        ```
        """))
        
    conn.close()
````
